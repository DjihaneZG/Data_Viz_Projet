```python
def count_rows(rows): 
    return len(rows)
```

```python
by_departement = df.groupby('code_departement').apply(count_rows)
by_departement.nlargest(15)
```

```python
#Quelles sont les dÃ©partements ayant le plus grand nombre de transactions immobiliÃ¨res ? 
plt.plot(by_departement.nlargest(15))
#ne pas oublier les label
```

```python
plt.figure(figsize = (30, 15))
plt.bar(range(1, 16), by_departement.nlargest(15))
plt.xticks(range(1, 16), by_departement.nlargest(15).index)
plt.xlabel('Departement')
plt.ylabel('Nombre de transaction immobiliÃ¨re')
plt.title('Frequency by DoM - Uber - April 2014')
```

```python
#Quelles sont les communes ou/et dÃ©partements ayant le plus grand nombre de transactions immobiliÃ¨res en 2020? (heat zone in France)
#plt.plot(by_departement)
plt.pie(by_departement.nlargest(6), explode = [0.4, 0.2, 0.1, 0, 0, 0])
plt.show() 
```
